```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.basketball-reference.com/wnba/boxscores/202406160IND.html'
```

```python
page = requests.get(url)
```

```python
soup = bs(page.text, 'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```
